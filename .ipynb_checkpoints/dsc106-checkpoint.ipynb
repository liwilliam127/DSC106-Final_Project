```python
path = '/Users/jg/ucsd/dsc106/DSC106-Final_Project'
url = '/Bengaluru_House_Data.csv'
data_url = path + url
house_df = pd.read_csv(data_url)
house_df = house_df.drop(columns=['society'])
house_df.dropna(inplace=True)
house_df.bath = house_df.bath.apply(lambda x: int(x))
house_df.balcony = house_df.balcony.apply(lambda x: int(x))
house_df.availability = house_df.availability.apply(\
                        lambda x: '20{}'.format(x[:2]) if x[:2].isnumeric() else x)
print(house_df.shape)
house_df.head()
```

```python
house_df.to_csv('house_clean.csv', index=False)
```
